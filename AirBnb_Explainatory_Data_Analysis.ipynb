**Understand the relationships between variables**

Chart - 01

```python
#chart - 1 vizzualization code
plt.scatter(df['price'],df['calculated_host_listings_count'],alpha=0.5)
plt.xlabel('Price')
plt.ylabel('Number of Airbnb Listings')
plt.title('Price vs Number of Airbnb Listings')
plt.show()
```

**1. Why did you pick the specific chart?**

The scatter plot with a regression line can be helpful in showcasing the relationship between price and Airbnb listings. The regression line gives a clear picture of how the price varies with an increase in the number of listings. If the regression line slopes upwards, it indicates that there is a positive correlation between the two variables, i.e., an increase in the number of listings leads to an decrease of number of listing

**2. What is/are the insight(s) found from the chart?**

It appears that there is a higher concentration of listings with lower prices, which may indicate that there is more demand for lower-priced listings.

**3. Will the gained insights help creating a positive business impact?**

Are there any insights that lead to negative growth? Justify with specific reason.

The insight that the number of listings is high in low pricing compared to high pricing can potentially have a positive impact on business. Airbnb hosts can consider setting lower prices for their listings to attract more customers and increase their occupancy rate. Additionally, lower prices can help in creating a competitive edge against other similar listings in the market.

However, this insight can also lead to negative growth if hosts consistently set prices too low, resulting in lower profit margins. Setting prices too low can also lead to a perception of lower quality and can attract less desirable guests. Therefore, hosts need to strike a balance between competitive pricing and maintaining profitability. Overall, while this insight can be useful, it needs to be analyzed in conjunction with other factors to make informed business decisions.

### **Chart - 02**

```python
#Chart - 2 visualization code
#create a new dataframe with the count of listings in each neighbourhood group
neighbourhood_count = df.groupby('neighbourhood_group').size().reset_index(name='count')
neighbourhood_count.value_counts()
```

**1. Why did you pick the specific chart?**

**2. What is/are the insight(s) found from the chart?**

The insight we can gather from this graph is that the two boroughs, Brooklyn and Manhattan, have a significantly higher number of Airbnb listings compared to the other three boroughs. This suggests that these two boroughs may be more popular among tourists or have more hosts who are willing to rent out their properties on Airbnb. This information can be useful for businesses looking to invest in the tourism industry or for hosts looking to start their own Airbnb listings.
